# Outliers detection


## [Isolation forest](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.IsolationForest.html?highlight=isolationforest)


```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import IsolationForest
from mlsummary.outliers import IsolationForestSummary

iris = load_iris()
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
X = iris.drop(columns=['target'])
y = iris['target']
iforest = IsolationForest().fit(X)
```

```python
iforest_summary = IsolationForestSummary(iforest, X = X)
iforest_summary.describe()
```

## [One Class support vector machine](https://scikit-learn.org/stable/modules/generated/sklearn.svm.OneClassSVM.html)

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.svm import OneClassSVM
from mlsummary.outliers import OneClassSVMSummary

iris = load_iris()
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
X = iris.drop(columns=['target'])
y = iris['target']
ocSVM = OneClassSVM().fit(X)
```

```python
ocSVM_summary = OneClassSVMSummary(ocSVM, X = X, store_X=True)
ocSVM_summary.describe()
```

```python
ocSVM_summary.plot()
```

## [Local Outlier Factor](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.LocalOutlierFactor.html)

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.neighbors import LocalOutlierFactor
from mlsummary.outliers import LocalOutlierFactorSummary

iris = load_iris()
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
X = iris.drop(columns=['target'])
y = iris['target']
lof = LocalOutlierFactor().fit(X)
```

```python
lof_summary = LocalOutlierFactorSummary(lof, X = X)
lof_summary.describe()
```

```python
lof_summary.plot(X=X)
```
