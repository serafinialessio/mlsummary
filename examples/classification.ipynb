## Linear discriminant analysis

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from mlsummary.classification import ldaSummary

iris = load_iris()
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
X = iris.drop(columns=['target'])
y = iris['target']
lda = LinearDiscriminantAnalysis().fit(X,y)
```

```python
lda_summary = ldaSummary(lda, X)
lda_summary.describe()
```

```python
lda_summary.plot(X = X, y = lda_summary.labels_pred)
```

## Classification Trees


```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from mlsummary.classification import TreeClassSummary


iris = load_iris()
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
X = iris.drop(columns=['target'])
y = iris['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 13)

tree = DecisionTreeClassifier().fit(X_train,y_train)
```

```python
tree_summary = TreeClassSummary(tree, X = X_test, y_true=y_test)
tree_summary.describe()
```

```python
tree_summary.plot_tree(feature_names = X_train.columns)
```

## Bagging classifier

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from mlsummary.classification import baggingClassSummary


iris = load_iris()
iris = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
X = iris.drop(columns=['target'])
y = iris['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 13)

bagging = BaggingClassifier().fit(X_train,y_train)
y_pred = bagging.predict(X_test)
```

```python
bagging_summary = baggingClassSummary(bagging, y_pred=y_pred, y_true=y_test)
bagging_summary.describe()
```

```python
bagging_summary.conf
```

```python
print(bagging_summary.ce)
print(bagging_summary.f1)
```
